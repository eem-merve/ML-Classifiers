```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("")
```

```python
data.shape
```

```python
data['PMTN'] = data['PMTN'].astype('int')
```

```python
grouped = data.groupby(['PMTN', 'Vector']).size().unstack(fill_value=0)

print(grouped)

grouped.plot(kind='bar', stacked=False)
plt.xlabel('PMTno')
plt.ylabel('Count')
plt.title('Distribution of 0s and 1s for each PMTno')
plt.legend(title='isCherenkov', labels=['0', '1'])
plt.show()
```

```python
grouped = data.groupby('PMTN').size()

grouped.plot(kind='bar')
plt.title('Data Counts For Each PMT')
plt.xlabel('PMTno')
plt.ylabel('Counts')
plt.xticks(rotation=0)
plt.show()
```

```python
data[data["Vector"] == 0]["PMTN"].hist(label="Vector=0")
data[data["Vector"] == 1]["PMTN"].hist(label="Vector=1")
plt.legend()
plt.xlabel("PMT Number")
plt.ylabel("Frequency")
plt.title("Distribution of PMTs for Vectors with 0 and 1")
plt.show()
```

```python
sns.lineplot(x="PMTN", y="Energy", data=data)
plt.xlabel("PMT Number")
plt.ylabel("Energy")
plt.title("Energy Distribution")
plt.show()
```

```python
sns.lineplot(x="PMTN", y="Time", data=data)
plt.xlabel("PMT Number")
plt.ylabel("Time")
plt.title("Time Distribution")
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(10, 6))

sns.lineplot(x="PMTN", y="Time", data=data, ax=ax1)

#second axis
ax2 = ax1.twinx()
sns.lineplot(x="PMTN", y="Energy", data=data, ax=ax2, color = "red")

ax1.set_xlabel("PMT Number")
ax1.set_ylabel("Time")
ax2.set_ylabel("Energy", color = "red")
plt.title("PMT Energy and Time Relationship")

plt.tight_layout()
plt.show()
```

```python
scintillation_data = data[data["Vector"] == 0]
```

```python
import matplotlib.pyplot as plt
sns.displot(scintillation_data, x="Energy", kind="kde")

plt.title("")

plt.show()
```

```python
sns.displot(scintillation_data, x="Time", kind="kde")

plt.title("Distribution of Scintillation Light Time Values")

plt.show()
```
